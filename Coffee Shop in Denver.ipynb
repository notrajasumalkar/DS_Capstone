```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
Denver_onehot = pd.get_dummies(Denver[['categories']], prefix="", prefix_sep="")

Denver_onehot['lat'] = Denver['lat'] 
Denver_onehot['lng'] = Denver['lng'] 

fixed_columns = [Denver_onehot.columns[-1]] + list(Denver_onehot.columns[:-1])
Denver_onehot = Denver_onehot[fixed_columns]

Denver_onehot.head()
```

```python
dnvr_grouped = Denver_onehot.groupby(['lat','lng']).mean().reset_index()
dnvr_grouped
```

# Using Kmeans Clustering to cluster the locations

```python
kclusters = 6

denver_grouped_clustering = dnvr_grouped.drop(['lat', 'lng'], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(denver_grouped_clustering)

kmeans.labels_[0:10]
```

```python
dnvr_grouped.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
dnvr_grouped
```

```python
dnvr_grouped.loc[dnvr_grouped['Cluster Labels'] == 1]
```
